```python
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
G = nx.Graph()
```

```python
urunler = ['Deodorant', 'Şampuan', 'Duş Jeli', 'Makarna', 'Meşrubat', 'Salça', 'Tuz', 'Bebek Bezi', 'Mama', 'Islak Mendil']
for urun in urunler:
    G.add_node(urun)
```

```python
G.add_edge('Deodorant', 'Şampuan', attr='M', color='blue')
G.add_edge('Deodorant', 'Duş Jeli', attr = 'M', color='blue')
G.add_edge('Şampuan', 'Duş Jeli', attr='M', color='blue')

G.add_edge('Deodorant', 'Makarna', attr='Y', color='green')
G.add_edge('Deodorant', 'Meşrubat', attr='Y', color='green')
G.add_edge('Duş Jeli', 'Makarna', attr='Y', color='green')
G.add_edge('Duş Jeli', 'Meşrubat', attr='Y', color='green')
G.add_edge('Makarna', 'Meşrubat', attr='Y', color='green')

G.add_edge('Salça', 'Makarna', attr='T', color='orange')
G.add_edge('Salça', 'Tuz', attr='T', color='orange')
G.add_edge('Salça', 'Bebek Bezi', attr='T', color='orange')
G.add_edge('Makarna', 'Tuz', attr='T', color='orange')
G.add_edge('Makarna', 'Bebek Bezi', attr='T', color='orange')
G.add_edge('Tuz', 'Bebek Bezi', attr='T', color='orange')

G.add_edge('Mama', 'Bebek Bezi', attr='S', color='gold')
G.add_edge('Mama', 'Islak Mendil', attr='S', color='gold')
G.add_edge('Bebek Bezi', 'Islak Mendil', attr='S', color='gold')
```

```python
np.random.seed(0)
pos = nx.spring_layout(G)
```

```python
plt.figure(figsize=(6, 6))
plt.axis("off")

cluster = dict(nx.algorithms.cluster.clustering(G))

nx.draw_networkx_nodes(G, pos, node_size=300, 
                       node_color='lightgray',
                       cmap=plt.cm.inferno)
nx.draw_networkx_edges(G, pos, alpha=0.500, edge_color='black')

nx.draw_networkx_labels(G, pos, font_color="black")

plt.show()
```

```python
plt.figure(figsize=(6, 6))
plt.axis("off")

cluster = dict(nx.algorithms.cluster.clustering(G))

nx.draw_networkx_nodes(G, pos, node_size=300, 
                       node_list=list(cluster.keys()),
                       node_color=list(cluster.values()),
                       cmap=plt.cm.inferno)
nx.draw_networkx_edges(G, pos, alpha=0.500, edge_color='lightgray')

nx.draw_networkx_labels(G, pos, font_color="black")

plt.show()
```

```python
plt.figure(figsize=(6, 6))
plt.axis("off")

degree_centrality = dict(nx.algorithms.centrality.degree_centrality(G))
nx.draw_networkx_nodes(G, pos, node_size=300, 
                       node_list=list(degree_centrality.keys()),
                       node_color=list(degree_centrality.values()),
                       cmap=plt.cm.inferno)
nx.draw_networkx_edges(G, pos, alpha=0.500, edge_color='lightgray')

nx.draw_networkx_labels(G, pos, font_color="black")

plt.show()
```

```python
plt.figure(figsize=(6, 6))
plt.axis("off")

closeness_centrality = dict(nx.algorithms.centrality.closeness_centrality(G))
nx.draw_networkx_nodes(G, pos, node_size=300, 
                       node_list=list(closeness_centrality.keys()),
                       node_color=list(closeness_centrality.values()),
                       cmap=plt.cm.inferno)
nx.draw_networkx_edges(G, pos, alpha=0.500, edge_color='lightgray')

nx.draw_networkx_labels(G, pos, font_color="black")

plt.show()
```

```python
plt.figure(figsize=(6, 6))
plt.axis("off")

betweenness_centrality = dict(nx.algorithms.centrality.betweenness_centrality(G))
nx.draw_networkx_nodes(G, pos, node_size=300, 
                       node_list=list(betweenness_centrality.keys()),
                       node_color=list(betweenness_centrality.values()),
                       cmap=plt.cm.inferno)
nx.draw_networkx_edges(G, pos, alpha=0.500, edge_color='lightgray')

nx.draw_networkx_labels(G, pos, font_color="black")

plt.show()
```

```python
plt.figure(figsize=(6, 6))
plt.axis("off")

eigenvector_centrality = dict(nx.algorithms.centrality.eigenvector_centrality(G))
nx.draw_networkx_nodes(G, pos, node_size=300, 
                       node_list=list(eigenvector_centrality.keys()),
                       node_color=list(eigenvector_centrality.values()),
                       cmap=plt.cm.inferno)
nx.draw_networkx_edges(G, pos, alpha=0.500, edge_color='lightgray')

nx.draw_networkx_labels(G, pos, font_color="black")

plt.show()
```

```python
df = {'degree': list(degree_centrality.values()),
      'cluster': list(cluster.values()),
      'closeness': list(closeness_centrality.values()),
      'betweenness': list(betweenness_centrality.values()),
      'eigenvector': list(eigenvector_centrality.values())}
df = pd.DataFrame(df, index=list(degree_centrality.keys()))
df.head(len(urunler))
```

```python
K = nx.karate_club_graph()
K.remove_edge(22,33)
np.random.seed(0)
pos = nx.fruchterman_reingold_layout(K)
```

```python
print("Number of nodes: ", K.order())
print("Number of edges: ", K.size())
```

```python
plt.figure(figsize=(8, 8))
plt.axis("off")

nx.draw_networkx_nodes(K, pos, node_size=300, 
                       node_color='black',
                       cmap=plt.cm.inferno)
nx.draw_networkx_edges(K, pos, alpha=0.500, edge_color='lightgray')

nx.draw_networkx_labels(K, pos, font_color="white")

plt.show()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 4))
degrees = dict(K.degree)
keys, values = zip(*sorted(degrees.items(), key=lambda x: -x[1]))
ax.stem(values)
ax.set_ylabel('Degree', fontsize=15)
ax.set_xlabel('Node ID', fontsize=15)
ax.set_xticks(range(len(keys)))
_ = ax.set_xticklabels(keys)
```

```python
plt.figure(figsize=(6, 6))
plt.axis("off")

degree_centrality = dict(nx.algorithms.centrality.degree_centrality(K))
nx.draw_networkx_nodes(K, pos, node_size=300, 
                       node_list=list(degree_centrality.keys()),
                       node_color=list(degree_centrality.values()),
                       cmap=plt.cm.inferno)
nx.draw_networkx_edges(K, pos, alpha=0.500, edge_color='lightgray')

nx.draw_networkx_labels(K, pos, font_color="black")

plt.show()
```
